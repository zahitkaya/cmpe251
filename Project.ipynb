```python
def is_public(name):
    return not (name.startswith('__') and name.endswith('__'))

def get_functions(source):
    return [name for name in dir(source) if callable(getattr(source, name)) and is_public(name)]

#print(f'Methods of class list: {get_functions(list)}', end='\n\n') kütüphanede hangi metodlar var onları gösteriyor.
#print(f'Functions of module math: {get_functions(math)}')
```

```python
import nltk
nltk.download("stopwords")
from nltk.corpus import stopwords
stp=stopwords.words('english')
df.insert(9,'outStop','')
```

```python
all_text=[[]]*100
for i in range (100):
    a=(df.description[i]+str(df.content[i])).lower().split()
    all_text[i]=[w for w in a if w not in stp]
    df.outStop[i]=all_text[i]
print(stopwords.words('english'))
```

```python
a=""
for i in range (100):

    fdist=nltk.FreqDist(all_text[i])
    b=fdist.most_common(1)
    df.freqDist[i]=b[0][0]
df   
```

```python
from difflib import SequenceMatcher
def similar (a,b):
   return SequenceMatcher(None,a,b).ratio() 
```

```python
similar (df.description[0],df.description[1])
```

```python
similarities = [["" for x in range(100)] for x in range(100)]
for i in range(100):
    for k in range (100):
        if (i!=k) : 
            similarities[i][k]=(similar(df.description[i],df.description[k]))
        if (i==k)  :
            similarities[i][k]=0  
        

        
```

```python
maximum = [0 for x in range(100)]
for i in range(100):
    for k in range (100):
        if(max(similarities[i])==similar(df.description[i],df.description[k])):
            maximum[i]=k

```
